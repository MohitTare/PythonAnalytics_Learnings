#Variable Transformation

```python
train.dtypes
```

```python
freq = train['Workclass'].value_counts()/train['Workclass'].shape[0]
```

```python
freq.dtype
```

```python
cat_var_list = list(train.dtypes.loc[train.dtypes == 'object'].index)
```

```python
cat_var_list
```

```python
for cat_var in cat_var_list:
    freq = train[cat_var].value_counts()/train[cat_var].shape[0]
    print freq
    categories_to_combine = list(freq.loc[freq < 0.05].index)
    print categories_to_combine
    for category in categories_to_combine:
        train[cat_var].replace({category:'Others'},inplace=True)
```

# Prediction via Skicit Learn

<h5> Lablel Encoding<h5>

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_var_list = list(train.dtypes.loc[train.dtypes == 'object'].index)
```

```python
cat_var_list
```

```python
le = LabelEncoder()
for var in cat_var_list:
    train[var] = le.fit_transform(train[var])
```

```python
train.dtypes
```

```python
from sklearn.tree import DecisionTreeClassifier
dependent_var = 'Income.Group'
independent_var = [x for x in train.columns if x not in ['ID',dependent_var]]
```

```python
independent_var
```

```python
model = DecisionTreeClassifier(max_depth=10,min_samples_leaf=100,max_features='sqrt')
```

```python
help(model)
```

```python
model.fit(train[independent_var],train[dependent_var])
```

```python
predictions_train = model.predict(train[independent_var])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_train = accuracy_score(train[dependent_var],predictions_train)
print acc_train
```
